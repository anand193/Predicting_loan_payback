```python
#lets see distribution of our target variable loan_paid_back
train['loan_paid_back'].value_counts().plot(kind = 'pie',autopct = '%0.2f%%')
plt.title('target_variable_distribution')
plt.show()
```

- this pie chart showing that there is 79.88% people had paid there loan while 20.12% have not. we have class imbalance in our dataset

```python
#lets see other categorical columns relation with our target variable 
t = 1
plt.figure(figsize = (10,12))
for i in train.select_dtypes(include = 'object'):
    pd.crosstab(train[i],train['loan_paid_back'],normalize = True).plot(kind = 'bar')
    plt.title(i)
    t += 1
plt.tight_layout()
plt.show()
```

```python
#lets see if our numerical columns have any relation with target variable
t = 1
plt.figure(figsize = (10,9))
for i in train.select_dtypes(include = 'number'):
    if i not in 'loan_paid_back':
        plt.subplot(3,3,t)
        sns.boxplot(y = train[i],x = train['loan_paid_back'])
        plt.title(i)
        t += 1
plt.tight_layout()
plt.show()
```

- people who had debt_to_income_ratio high and interest_rate high they are the one who could not pay back their loan

```python
#lets see if there any correlation between variables
sns.heatmap(train.select_dtypes(include= 'number').corr(),vmin = -1,vmax = 1,annot = True,linecolor='yellow')
plt.title('heatmap')
plt.show()
```

```python
train['grade_subgrade'].unique()
```

```python
#lets convert our categorica columns into nummerical columns 
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in train.select_dtypes(include = 'object'):
    if i not in ['grade_subgrade','education_level']:
        train[i] = encoder.fit_transform(train[i])
```

```python
train['education_level'].unique()
```

```python
#since there is an order in education level we will do ordinal encoding here
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories = [['High School', "Master's", "Bachelor's", 'PhD', 'Other']])
train['education_level'] = encoder.fit_transform(train[['education_level']])
```

```python
train['education_level'].unique()
```

```python
#lets seperate our data into x and y for splitting it into train and validation 
x = train.drop('loan_paid_back',axis = 1)
y = train['loan_paid_back']
```

```python
#lets drop our grade_subgrade and id column columns since there are not usable in predicting loan amount
x.drop(['id','grade_subgrade'],axis = 1,inplace = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,roc_curve,roc_auc_score
```

```python
x_train,x_val,y_train,v_val = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

```python
#base model
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model_gb = GradientBoostingClassifier()
model_gb.fit(x_train,y_train)
y_pred = model_gb.predict(x_val)
```

```python
print(classification_report(v_val,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(v_val,y_pred))
```

```python
y_pred_proba = model_gb.predict_proba(x_val)[:, 1]
```

```python
print(roc_auc_score(v_val,y_pred_proba))
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
y_pred_proba = model_gb.predict_proba(x_val)[:, 1]

fpr, tpr, thresholds = roc_curve(v_val, y_pred_proba)
roc_auc = auc(fpr, tpr)
print("ROC-AUC Score:", roc_auc)

# Step 4: Plot ROC Curve
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--') 
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-AUC Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
#lets test other models 
from xgboost import XGBClassifier
```

```python
model_xg = XGBClassifier()
model_xg.fit(x_train,y_train)
y_pred = model_xg.predict(x_val)
y_pred_proba = model_xg.predict_proba(x_val)[:, 1]
print(f'classification_report: {classification_report(v_val,y_pred)}')
print(f'confusion_matrix: {confusion_matrix(v_val,y_pred)}')
print(f'roc_auc_score: {roc_auc_score(v_val,y_pred_proba)}')
```

```python
model_xg = XGBClassifier(n_estimators=150,max_depth = 9,learning_rate= 0.1)
model_xg.fit(x_train,y_train)
y_pred = model_xg.predict(x_val)
y_pred_proba = model_xg.predict_proba(x_val)[:, 1]
print(f'classification_report: {classification_report(v_val,y_pred)}')
print(f'confusion_matrix: {confusion_matrix(v_val,y_pred)}')
print(f'roc_auc_score: {roc_auc_score(v_val,y_pred_proba)}')
```

```python
#lets train neural network to see if we can improve our accuracy
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential([
        Dense(32,activation = 'relu',input_shape = (x_train.shape[1],)),
        Dense(64,activation = 'relu',input_shape = (x_train.shape[1],)),
        Dense(1,activation = 'sigmoid')
])
model.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
history = model.fit(x_train,y_train,epochs = 10,validation_split = 0.2,batch_size = 16)
```

```python
loss,accuracy = model.evaluate(x_val,v_val)
print(accuracy)
```

```python
plt.plot(history.history['accuracy'],label ='train_accuracy')
plt.plot(history.history['val_accuracy'],label = 'val_accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```
